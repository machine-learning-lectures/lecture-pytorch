# 🟢 Solution: Implement ReLU

Reference solution for the ReLU activation function.

$$\text{ReLU}(x) = \max(0, x)$$


```python
import torch
```

```python
# ✅ SOLUTION

def relu(x: torch.Tensor) -> torch.Tensor:
    return x * (x > 0).float()
```

```python
# Verify
x = torch.tensor([-2., -1., 0., 1., 2.])
print("Input: ", x)
print("Output:", relu(x))
```
